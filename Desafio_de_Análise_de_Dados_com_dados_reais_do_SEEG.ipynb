```python
import pandas as pd
```

```python
url = '/content/drive/MyDrive/Alura_data_science/1-SEEG10_GERAL-BR_UF_2022.10.27-FINAL-SITE.xlsx'
#IMPORTAÇÃO DO LINK DA PLANILHA DE EXCEL SEEG
```

```python
dados_bruto_seeg = pd.read_excel(url, sheet_name= 'GEE Estados')

#TRABALHANDO COM A PLANILHA GEE ESTADOS
```

```python
dados_bruto_seeg
```

```python
dados_bruto_seeg.loc[:,'Nível 1 - Setor':'Produto']
```

```python
info_colunas = list(dados_bruto_seeg.loc[:,'Nível 1 - Setor':'Produto'].columns)
info_colunas
```

```python
dados_bruto_seeg.loc[:,1970:2021]
```

```python
ano = list(dados_bruto_seeg.loc[:,1970:2021].columns)
ano
```

```python
dados_novos_SEEG = dados_bruto_seeg.melt(id_vars= info_colunas, value_vars= ano, value_name='Emissão', var_name='Ano')
dados_novos_SEEG
#MEU OBJETIVO FOI CRIAR UMA BASE DE DADOS TRANSFORMANDO OS ANOS(COLUNAS EM LINHAS)
```

```python
#ATIVIDADE 1

agrup_dados = dados_novos_SEEG.groupby(['Estado', 'Nível 1 - Setor'])['Emissão'].sum()

#OBTÉM UMA TABELA AGRUPANDO ESTADOS, NÍVEL 1 E SOMA AS EMISSÕES DAS ATIVIDADES
#Variavel Agrup_dados : Agrupa por Estado, Nivel 1 e soma as emissões
agrup_dados.head(60)
```

```python
#ATIVIDADE 3

agrup_dados.loc['MG'].idxmax()

#PEGA O SETOR DO MAIOR VALOR DO ESTADO DE MINAS GERAIS
```

```python
#ATIVIDADE 4
max_emissao = agrup_dados.groupby(level=1).idxmax()

#TRANSFORMA EM TABELA A MAIOR ATIVIDADE ECONOMICA POR ESTADO
```

```python
#ATIVIDADE 5
max_emissao_Estado = agrup_dados.groupby(level=0).idxmax()

#OBTÉM UMA TABELA CONTENDO A MÁXIMA EMISSÃO PARA CADA ATIVIDADE ECONÔMICA
```
